```python
import matplotlib.pyplot as plt
import numpy as np

#To draw a line of ax+by+c=0

#Line1
a=3
b=4
c=0
x=np.arange(-5,5)
plt.plot(x,x*(-a/b)+c)

#Line2
a=3
b=4
c=-4
plt.plot(x,x*(-a/b)+c)

#Line3
a=5
b=3
c=0
plt.plot(x,x*(-a/b)+c)

plt.legend(['3x+4y+0=0','3x+4y-4=0','5x+3y=0'],loc='upper left')

plt.show()
```
